# **Correlation analysis between numerical columns.**

```python
# Select only the numerical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64'])
# Calculate the correlation matrix
correlation_matrix = numerical_columns.corr()

# Set the figure size
plt.figure(figsize=(10, 8))

# Create a heatmap to visualize the correlation matrix
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix of Numerical Columns')

# Display the plot
plt.show()
```

**Analysis**: By analysing the output we can conclude that

*   **Sales and Profit (0.48):** It indicates a moderately strong positive correlation between 'Sales' and 'Profit'. This means that as sales increase, profit also tends to increase. This is a typical and expected relationship in business, as higher sales often lead to higher profits.
*  **Sales and Quantity (0.20):** It indicates a weak positive correlation between 'Sales' and 'Quantity'. This suggests that higher sales are somewhat associated with a larger quantity of items sold, but the relationship is not very strong.
*   **Sales and Discount (-0.028):** It indicates there is a very weak negative correlation between 'Sales' and 'Discount', indicating that discounts slightly reduce sales.
*   **Quantity and Profit (0.066):** Indicates a very weak positive correlation between 'Quantity' and 'Profit'. This weak correlation might be because profit margins vary across products.
*   **Quantity and Discount (0.0087):** There is an almost negligible positive correlation between 'Quantity' and 'Discount'.This could mean that discounts are either not substantial enough to affect purchasing behavior or that they are applied to low-quantity items.

*   **Discount and Profit (-0.22):** There is a weak negative correlation between 'Discount' and 'Profit'. This means that higher discounts are generally associated with lower profits, which is expected because discounts reduce the selling price and hence the profit margin.

**Conclusions:**

1. **Sales and Profit:** The strongest relationship observed is between sales and profit. This reinforces the importance of increasing sales to drive profit growth.

2. **Discount Impact:** Discounts have a notable negative impact on profit and a very weak negative impact on sales, suggesting that discounts might not be the best strategy for driving sales and could harm profitability.

3. **Quantity Relationships:** The weak correlations involving quantity suggest that other factors (like product pricing and cost structure) play a significant role in driving sales and profit.

# **Scatter plot to analyze the relationship between sales and profit**

```python
#scatter plot to analyze the relationship between sales and profit.

if 'Sales' in df.columns and 'Profit' in df.columns: #Check if 'Sales' and 'Profit' column exsists in the dataframe
  plt.figure(figsize=(10,6)) #Set the figure size
  sns.regplot(x='Sales',y='Profit',data=df,color='orange') #Plot a scatter plot with regression line
  plt.xlabel('Sales') #Set the x-axis label
  plt.ylabel('Profit') #Set the y-axis label
  plt.title('Scatter plot of Sales vs Profit') #Set the title
  plt.grid(True) #Add a grid to the plot
  plt.xticks(rotation=45) #Rotate the x-axis labels by 45 degrees for better readability
  plt.show() #Display the plot
```


**Analysis:** By analysing the output we can conclude that,

1. **Correlation**:There is a positive correlation between sales and profit, meaning that as sales increase, profit tends to increase as well. This relationship is indicated by the upward trend of the data points and the fitted line.

2. **Variability**:  At lower sales values, the points are densely packed, showing less variability in profit. As sales increase, the variability in profit also increases.

3. **Outliers**:There are some outliers where sales are high, but profit is either very low or even negative. These outliers indicate instances where high sales did not translate to high profits, which could be due to high costs, discounts, or other factors.

4. **Regression Line**:The fitted line and the shaded confidence interval provide a visual of the general trend and the uncertainty around it. The positive slope of the regression line confirms the positive correlation between sales and profit.

**Insights:**

1.**Sales Impact on Profit**: Generally, increasing sales leads to increased profits. However, the relationship is not perfectly linear, indicating other factors may influence profitability.

2.**High Variability at Higher Sales**: The higher variability at larger sales values suggests that profit margins may vary significantly for larger sales. This could be due to variable costs, discounts, or different types of products/services sold.

3.**Negative Profits**:The presence of data points with negative profits, even at higher sales levels, highlights the need to investigate the causes of these losses.

**Strategic Implications:**

 1. **Improving Profit Margins**:Investigate the factors contributing to variability in profits, especially at higher sales levels. This could involve analyzing cost structures, pricing strategies, and discount policies.
 2. **Addressing Loss-Making Sales**: Analyze the instances where high sales result in low or negative profits. Identifying patterns or common factors in these cases can help in formulating strategies to mitigate such issues.

 3. **Optimizing Sales Strategies**:Given the positive correlation between sales and profit, strategies to boost sales could be beneficial. However, it is crucial to ensure that these strategies do not compromise profit margins.

 4. **Focus on Outliers**:The outliers where sales are high but profits are low or negative should be a focus for detailed analysis. Understanding these exceptions can provide valuable insights for improving overall profitability

5. **Segment Analysis**: Break down the data by different segments (e.g., region, product type) to see if the correlation holds across all segments or if there are significant differences.
6. **Cost Analysis**:Analyze the cost components associated with high sales but low/negative profit instances to identify areas for cost optimization.
7. **Time Series Analysis**:Time series analysis to identify trends, seasonality, and cyclical patterns in sales and profit.
